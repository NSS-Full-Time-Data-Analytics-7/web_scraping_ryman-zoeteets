## find openers

```python
ryman_shows = soup.findAll('div', attrs={'class':'tribe-beside-image'})
ryman_shows
```

```python
openers = []
for x in ryman_shows:
    if re.search("with.+", str(x.findAll('span'))):
        openers.append(x.findAll("span")[-1].text)
    elif re.search("featuring.+", str(x.findAll('span'))):
        openers.append(x.findAll('span')[-1].text)
    else:
        openers.append('no opener')
openers
```

```python
openers = pd.DataFrame(openers)
openers = openers.rename(columns = {0: 'opener'})
openers
```

## find sponsors

```python
sponsors = []
for x in ryman_shows:
    if re.search("Tour.+", str(x.findAll('span'))):
        sponsors.append(x.find("span").text)
    elif re.search("at the.+", str(x.findAll('span'))):
        sponsors.append(x.find('span').text)
    elif re.search("Benefiting.+", str(x.findAll('span'))):
        sponsors.append(x.find('span').text)
    else:
        sponsors.append('no sponsor')
sponsors
```

```python
sponsors = pd.DataFrame(sponsors)
sponsors = sponsors.rename(columns = {0:'sponsor'})
sponsors
```

## Headliners, show time, show day, opener, sponsor

```python
headliners['opener'] = openers.opener
headliners['sponsor'] = sponsors.sponsor

headliners_page1 = headliners
headliners_page1
```
